```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
books = []
# There are a total of 50 pages so I created a loop for all the pages to extract data
for i in range(1,51):
  url =f"https://books.toscrape.com/catalogue/page-{i}.html"
  response  = requests.get(url)
  soup = BeautifulSoup(response.content,'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article',class_='product_pod')

    
  for article in articles:
    # We are getting the names from the image title class
    image = article.find('img')
    title = image.attrs['alt']
    star = article.find('p')
    star = star['class'][1]
    price = article.find('p',class_='price_color').text
    price = float(price[1:])
    books.append([title,price,star])
```

```python
books_df = pd.DataFrame(books,columns=['Title','Price','StarRating'])
# Since it is a list of lists we create columns for each
```

```python
books_df.to_csv('Books.csv')
```
